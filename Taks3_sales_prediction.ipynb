```python
#Seperating the data for Train and Test


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(sales_data[['TV']],sales_data[['Sales']],test_size=0.2,random_state=0) 
```

```python
#Linear Regression for training the mmodel


from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python


#prediction of TV ADS


repos=model.predict(x_test)
print(repos)
```
